```python
### At least have some information of Classes.
### To create classes, use 'class' and initialize class attributes by method called __init__
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        
### To create instances of Rectangle class, call class name with arguments which passed to __init__ method.
### First arguments like self is automatically filled in by python and contains the object being created.
### self is convention, call it whatever name you choose.

r1 = Rectangle(10, 30)
r2 = Rectangle(100, 200)

print(r1.height)
print(r2.width)

print('===================')

### Attributes which are callables are methods, if not properties
### So, height and width are properties
### get properties and methods by dot notation.
### when create instance methods, methods require first argument to be the object being used when the method is called.

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(the_referenced_object):
        return 2 * (the_referenced_object.width + the_referenced_object.height)
    
r1 = Rectangle(5, 20)
print(r1.area())
print(r1.perimeter())

### As you can see above, r1 is object. when area is called, python called area method passing r1 to self parameter
### Also using another name instead of 'self' is available

### Python defines a bunch of special methods that can use to give our classes functionality 
### that resembles functionality of built-in and standard library objects
### These special methods provide us an easy way to overload operators in python
### For example, we can obtain the string representation of an integer using the built-in str function
print('================')
print(str(10))
print(r1)

### display the rectangle as string

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(the_referenced_object):
        return 2 * (the_referenced_object.width + the_referenced_object.height)
    def to_str(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)
    
r1 = Rectangle(10, 20)
r1.to_str()

### when call str method, python will find Recteangle class and method called __str__ 
### If exist, return that value
### cf) __repr__ exist
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(the_referenced_object):
        return 2 * (the_referenced_object.width + the_referenced_object.height)
    def __str__(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)
print('=================')    
r1 = Rectangle(1, 4)
print(str(r1))
print(r1)


### still, python doesn't convert r1 into string, So __repr__ method needs
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(the_referenced_object):
        return 2 * (the_referenced_object.width + the_referenced_object.height)
    def __str__(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)
    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)
print('=================')    
r1 = Rectangle(1, 4)
print(r1) ### __str__
r1 ### __repr__

### Let's see something equal size
r1 = Rectangle(10, 10)
r2 = Rectangle(10, 10)
print('==================')
if r1 == r2:
    print('same')
else:
    print('not same')
    
### why not same cuz of same size?
### look into __eq__ method

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(the_referenced_object):
        return 2 * (the_referenced_object.width + the_referenced_object.height)
    def __str__(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)
    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)
    def __eq__(self, other):
        print('self={0}, other={1}'.format(self, other))
        if isinstance(other, Rectangle):
            return (self.width, self.height) == (other.width, other.height)
        else:
            return False

print('=================')
r1 = Rectangle(10, 10)
r2 = Rectangle(10, 10)
r1 is r2 ### not same memory address
r1 == r2
r3 = Rectangle(1, 2)
r1 == r3
r1 == 100

### After then, remove a print line
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(the_referenced_object):
        return 2 * (the_referenced_object.width + the_referenced_object.height)
    def __str__(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)
    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)
    def __eq__(self, other):
        if isinstance(other, Rectangle):
            return (self.width, self.height) == (other.width, other.height)
        else:
            return False
        
### Let's see <, >, = etc..
### Python has special method to use them which is __lt__, __gt__, __le__, etc..
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(the_referenced_object):
        return 2 * (the_referenced_object.width + the_referenced_object.height)
    def __str__(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)
    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)
    def __eq__(self, other):
        if isinstance(other, Rectangle):
            return (self.width, self.height) == (other.width, other.height)
        else:
            return False
    def __lt__(self, other):
        if isinstance(other, Rectangle):
            return self.area() < other.area()
        else:
            return NotImplemented
r1 = Rectangle(10, 10)
r2 = Rectangle(20, 20)
print('==================')
print(r1 < r2)
print(r2 < r1)
print(r1 > r2) ### might be confused cuz there is no such method '>' in class
### Here is a thing. Python is clever so 'r1 > r2' is converted into 'r2 < r1'

### print(r1 >= r2) 
### cause error, no such thing 'greater than "__gt__" in class'

### Back to first scenario, simple version
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def __repr__(self):
        return 'Rectangle {0}, {1}'.format(self.width, self.height)
r1 = Rectangle(10, 20)
print('=============')
r1.width  ### 10
r1.width = 20
r1  ### 20, 20

### As you can see, get and set width property directly
### And get width like -2 would be ridiculous, So we need Somthing to do
### Called Decorators to encapsulate property getters and setters
class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height
    def __repr__(self):
        return 'Rectangle {0}, {1}'.format(self.width, self.height)

    @property
    def width(self):
        return self._width
    
    @width.setter
    def width(self, width):
        if width <= 0:
            raise ValueError('Width must greater than 0')
        self._width = width
        
    @property
    def height(self):
        return self._height
    
    @height.setter
    def height(self, height):
        if height <= 0:
            raise ValueError('Height must greater than 0')
        self._height = height
        
r1 = Rectangle(10, 20)
print('=============')
print(r1.width)
r1

### raise error 'r1.width = -10'
### But still make instance a rectangle which size is negative
### So, we must check width, height value in __init__ phase.
### Using the accessor methods for width and height

class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height
        ### now we call our accessor methods to set the width and height
        self.width = width
        self.height = height
        
    def __repr__(self):
        return 'Rectangle {0}, {1}'.format(self.width, self.height)

    @property
    def width(self):
        return self._width
    
    @width.setter
    def width(self, width):
        if width <= 0:
            raise ValueError('Width must greater than 0')
        self._width = width
        
    @property
    def height(self):
        return self._height
    
    @height.setter
    def height(self, height):
        if height <= 0:
            raise ValueError('Height must greater than 0')
        self._height = height
        
          
# r1 = Rectangle(10, -10) makes error
### Still many methods needed to check value whether these are numeric type and so on.
### Of course in __init_ phase
### when created is better than when methods are called such as area, and raise an exeception. For Debugging
```
